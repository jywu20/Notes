```python
with open('d:/Notes/Computer Science/Scientific computation with Python/example.csv') as fp:
    f_csv = csv.reader(fp)
    header = next(f_csv)
    Row = namedtuple('Row', header)
    # Original data are strings
    for item in f_csv:
        row = Row(*map(float, item))
        print(row)
        
```
